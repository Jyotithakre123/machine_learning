```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = ps.read_csv("C:\\Users\\Sukhmeetkour\\OneDrive\\Documents\\Salary Data.csv")
df
```

```python
new_df = df.drop(['Age', 'Gender', 'Education Level', 'Job Title'], axis=1)
new_df
                  
```

```python
#find the null values in a dataset
new_df.isnull().sum()
```

```python
#fill null values with median values
new_df['Salary'].fillna(new_df['Salary'].median(),inplace=True)
new_df['Years of Experience'].fillna(new_df['Years of Experience'].median(),inplace=True)
```

```python
new_df.isnull().sum()
```

```python
#split into dependent and independent
x = np.array(new_df['Years of Experience']).reshape(-1, 1)  # independent variable
y = np.array(new_df['Salary'])              # dependent variable

```

```python
x
```

```python
y
```

```python
# split the dataset into training and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#model building
```

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(x_train, y_train)
```

```python
#model prediction
```

```python
y_test
```

```python
y_pred = LinReg.predict(x_test)
y_pred
```

```python
act_pred_df = ps.DataFrame({'Actual':y_test,
                            "Predicted":y_pred})
act_pred_df
```

#Model accuracy

```python
LinReg.score(x_train,y_train)*100
```

```python
LinReg.score(x_test,y_test)*100

```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
