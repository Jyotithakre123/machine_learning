```python
box= soup.find_all('div',class_ = 'tUxRFH')

box
```

```python
product_name=[]
storage_info=[]
camera_info=[]
batry_info=[]
warrenty_name=[]
price_info=[]



for i in box:
    product=i.find("div",class_="KzDlHZ")
    detail=i.find_all("li",class_="J+igdf")
    price=i.find("li",class_="Nx9bqj _4b5DiR")
    product_name.append(product.text.strip())
    
    storage_info.append(detail[0].text.strip() if len(detail)>0 else"")
    camera_info.append(detail[1].text.strip() if len(detail)>2 else"")
    batry_info.append(detail[2].text.strip() if len(detail)>3 else"")
    warrenty_name.append(detail[3].text.strip() if len(detail)>4 else"")

    
    price_info.append(product.text.strip())
    
    
print(product_name)
print(storage_info)
print(camera_info)
print(batry_info)
print(warrenty_name)
print(price_info)
```
