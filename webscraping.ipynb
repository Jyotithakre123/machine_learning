```python
# Import required library
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
Response = requests.get(url)
Response    
```

```python
Response.content
```

```python
soup = BeautifulSoup(Response.text,'html.parser')
print(soup.prettify())
#soup
```

```python
# Extract the title
name = soup.find_all('a',class_ = 'title')
name_list = []
for i in name:
    title = i.text
    name_list.append(title.strip())
print(name_list)
```

```python
# Extract the price

price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Price Extraction
price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Extract the description
desc = soup.find_all('p',class_ = 'description card-text')
desc_list = []
for i in desc:
    description = i.text
    desc_list.append(description.strip())
print(desc_list)
```

```python
# Extract Review

review = soup.find_all('p',class_ = 'review-count float-end')
review_list = []
for i in review:
    review_ = i.text
    review_list.append(review_.strip())
print(review_list)
```

```python
# convert this information into dataframe

df = pd.DataFrame({
    "Title":name_list,
    "Price":price_list,
    "Description":desc_list,
    "Review":review_list
})
```

```python
product_name = []
storage_info = []
camera_info = []
batry_info = []
warrenty_name = []
price_info = []

for i in Box:
    # Extract product name
    product = i.find('div', class_='KzDlHZ')
    details = i.find_all('li', class_='J+igdf')
    price = i.find('div', class_='Nx9bqj _4b5DiR')
    
    product_name.append(product.text.strip())
    # Extract all li tags with class 'J+igdf'
    details = i.find_all('li', class_='J+igdf')
    storage_info.append(details[0].text.strip() if len(details) > 0 else "")
    camera_info.append(details[1].text.strip() if len(details) > 1 else "")
    batry_info.append(details[2].text.strip() if len(details) > 2 else "")
    warrenty_name.append(details[3].text.strip() if len(details) > 3 else "")

    # Extract price
    price = i.find('div', class_='Nx9bqj _4b5DiR')
    price_info.append(price.text.strip() if price else "")

print(product_name)
print(storage_info)
print(camera_info)
print(batry_info)
print(warrenty_name)
print(price_info)
    
```
