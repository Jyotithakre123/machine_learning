```python
def remove_outliers(House, column):
    Q1 = House[column].quantile(0.25)
    Q3 = House[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return House[(House[column] >= lower) & (House[column] <= upper)]

df_filtered = df.copy()
for col in ['price', 'sqft_living', 'bedrooms', 'bathrooms']:
    df_filtered = remove_outliers(df_filtered, col)

print("Filtered dataset shape:", df_filtered.shape)
```

```python
# Area vs Price
plt.scatter(df_filtered['sqft_living'], df_filtered['price'])
plt.xlabel("Living Area (sqft)")
plt.ylabel("Price")
plt.title("Area vs House Price (Outliers Removed)")
plt.show()

# Bedrooms vs Price
sns.boxplot(x='bedrooms', y='price', data=df_filtered)
plt.title("Bedrooms vs Price (Outliers Removed)")
plt.show()

# Bathrooms vs Price
sns.boxplot(x='bathrooms', y='price', data=df_filtered)
plt.title("Bathrooms vs Price (Outliers Removed)")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Encode binary columns
le = LabelEncoder()
df['furnishing_status'] = le.fit_transform(df['furnishing_status'])
df['mainroad'] = le.fit_transform(df['mainroad'])

# Encode multi-category columns
Lab_en = LabelEncoder()
cat_col = ["street", "city", "statezip", "country"]
for col in cat_col:
    df[col] = Lab_en.fit_transform(df[col])

# Define X and y
x = df.drop("price", axis=1)
y = df["price"]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

# Linear Regression
LinReg = LinearRegression()
LinReg.fit(x_train, y_train)
```
